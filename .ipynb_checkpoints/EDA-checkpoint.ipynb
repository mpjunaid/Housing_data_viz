Checking the datatype of each columns and the missing values count

```python
df.info()
```

It can be seen that the there is only one column "total_bedrooms" is having missing values 

Since out of all 20640 there is only 207 values that are missing 

The rows are generally seen dropped and i have just analyzing these values hence created a data frame called 'df_missing'

```python
df_missing=df[df.isnull().any(axis=1)]
```

```python
df_missing.info()
```

```python
df_missing=df_missing.drop(columns=["total_bedrooms"])
```

```python
df_missing.info()
```

```python
df_missing.describe()
```

```python
df.describe()
```

```python
df_data=df_missing
len(df_data.columns)
```

```python
df_data.columns
```

```python
df_data=df_missing
fig, axs = plt.subplots(len(df_data.columns), len(df_data.columns), figsize=(12,12))
cc = ['blue','orange']
bb = df_data.columns

for i in range(len(df_data.columns)):
    for j in range(len(df_data.columns)):
        axs[i,j].scatter(x=df_data[bb[j]],y=df_data[bb[i]],color=cc[0],s=1)
        if j==0:
            axs[j,i].set_title(bb[i], fontsize=8)
        if i==0:
            axs[j,i].set_ylabel(bb[j], fontsize=8)
        axs[i,j].tick_params(axis='x', labelsize=8)
        axs[i,j].tick_params(axis='y', labelsize=8)
plt.show()
```

The following plot shows the place where the values are missing their values 


The missing values are in orange domain

```python
df_data=df.drop(columns=["total_bedrooms"])
fig, axs = plt.subplots(len(df_data.columns), len(df_data.columns), figsize=(12,12))
cc = ['blue','orange']
bb = df_data.columns

for i in range(len(df_data.columns)):
    for j in range(len(df_data.columns)):
        axs[i,j].scatter(x=df_data[bb[j]],y=df_data[bb[i]],color=cc[0],s=1)
        axs[i,j].scatter(x=df_missing[bb[j]],y=df_missing[bb[i]],color=cc[1],s=1)
        if j==0:
            axs[j,i].set_title(bb[i], fontsize=8)
        if i==0:
            axs[j,i].set_ylabel(bb[j], fontsize=8)
        axs[i,j].tick_params(axis='x', labelsize=5)
        axs[i,j].tick_params(axis='y', labelsize=5)
plt.show()
```
